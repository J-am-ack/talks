
##### 1. 编程基础模块
###### 1.1 变量与基本数据类型
```python
# 演示：变量定义+数据类型查看
num_int = 10  # int类型
num_float = 3.14  # float类型
is_true = True  # bool类型（属于int子类）

# 查看类型
print("num_int类型：", type(num_int))
print("num_float类型：", type(num_float))
print("is_true类型：", type(is_true))
```


###### 1.2 运算符与表达式
```python
# 演示：算术/关系/逻辑运算符
a = 10
b = 3

# 算术运算
print("加法：", a + b)
print("除法（浮点数）：", a / b)
print("取整除法：", a // b)
print("取余：", a % b)

# 关系运算
print("a > b？", a > b)

# 逻辑运算（结合关系运算）
print("a>5 且 b<5？", (a>5) and (b<5))
```


###### 1.3 控制结构
```python
# 演示1：分支结构（if-elif-else）
score = 85
if score >= 90:
    print("优秀")
elif score >= 80:
    print("良好")
else:
    print("继续努力")


# 演示2：循环结构（for+while）
# for循环遍历列表
print("\nfor循环遍历：")
for i in [1,2,3,4]:
    print(i, end=" ")

# while循环计数
print("\n\nwhile循环计数：")
count = 0
while count < 3:
    print(f"计数：{count}")
    count += 1
```


##### 2. 函数模块
```python
# 演示：函数定义+参数+返回值
def calculate_sum(a, b=0):  # b是默认参数
    """计算两个数的和（默认第二个数为0）"""
    return a + b

# 调用函数
print("调用1（传1个参数）：", calculate_sum(5))
print("调用2（传2个参数）：", calculate_sum(5, 3))
```


##### 3. 容器类型模块
```python
# 3.1 字符串str
s = "hello python"
print("字符串切片：", s[0:5])  # 取前5个字符
print("字符串分割：", s.split(" "))  # 按空格分割


# 3.2 列表list
lst = [1,2,3,4]
lst.append(5)  # 新增元素
print("列表新增后：", lst)
print("列表弹出最后一个元素：", lst.pop())


# 3.3 字典dict
student = {"name": "小明", "age": 18}
print("字典取值：", student["name"])
student["score"] = 90  # 新增键值对
print("字典新增后：", student)


# 3.4 集合set（去重）
s1 = {1,2,2,3}
s2 = {3,4,5}
print("集合去重后：", s1)
print("集合交集：", s1 & s2)
```


##### 4. 问题求解与算法
```python
# 演示：简单冒泡排序（算法基础）
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]  # 交换
    return arr

# 测试
test_arr = [5,2,9,1]
print("排序前：", test_arr)
print("排序后：", bubble_sort(test_arr))
```


##### 5. 面向对象与数据结构初步
```python
# 5.1 面向对象（类与对象）
class Student:
    def __init__(self, name, age):  # 构造方法
        self.name = name
        self.age = age
    
    def introduce(self):  # 成员方法
        print(f"我是{self.name}，今年{self.age}岁")

# 创建对象
stu = Student("小红", 19)
stu.introduce()


# 5.2 数据结构初步（列表模拟栈）
class Stack:
    def __init__(self):
        self.stack = []
    
    def push(self, item):  # 入栈
        self.stack.append(item)
    
    def pop(self):  # 出栈
        return self.stack.pop() if self.stack else "栈为空"

# 测试栈
s = Stack()
s.push("A")
s.push("B")
print("栈出栈元素：", s.pop())
```
